**Data set from https://www.kaggle.com/competitions/playground-series-s3e14**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv("train.csv")
```

```python
test_data = pd.read_csv("test.csv")
```

```python
train_data.head()
```

```python
X = train_data.drop(columns = ["id","yield"],axis = 1)
y = train_data["yield"]
```

```python
Xt = test_data.drop(columns = ["id"],axis = 1)
```

```python
X.info()
```

```python
X = X.dropna()
```

```python
X.head()
```

```python
X.corr()
```

```python
X.drop(columns = ["MinOfUpperTRange","MinOfUpperTRange","MaxOfLowerTRange","MinOfLowerTRange","RainingDays","MaxOfUpperTRange"],axis = 1,inplace=True)
```

```python
Xt.drop(columns = ["MinOfUpperTRange","MinOfUpperTRange","MaxOfLowerTRange","MinOfLowerTRange","RainingDays","MaxOfUpperTRange"],axis = 1,inplace=True)
```

```python
X.head()
```

```python
Xt.head()
```

```python
X["NewAverage"] = X["AverageOfUpperTRange"] +  X["AverageOfLowerTRange"]
```

```python
Xt["NewAverage"] = Xt["AverageOfUpperTRange"] +  Xt["AverageOfLowerTRange"]
```

```python
X = X.drop(columns = ["AverageOfLowerTRange","AverageOfLowerTRange"])
```

```python
Xt = Xt.drop(columns = ["AverageOfLowerTRange","AverageOfLowerTRange"])
```

```python
X["clonesize"].value_counts()
```

```python
cols = X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=58)

from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0,1))

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

X_train = pd.DataFrame(data = X_train,columns=[cols])
X_test = pd.DataFrame(data = X_test,columns=[cols])
```

```python
Xt = sc.transform(Xt)
Xt = pd.DataFrame(data = Xt,columns=[cols])
```

```python
from sklearn.svm import SVR
svr = SVR(C=2500,max_iter=500,gamma=0.04)
svr.fit(X_train,y_train)
prediction = svr.predict(X_test)

from sklearn.metrics import mean_absolute_error
print(f"Mean Absolute Error = {str(mean_absolute_error(y_test,prediction))}")
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(n_estimators=425,max_depth=5,learning_rate=0.020, random_state=58)

xgb.fit(X_train,y_train)
predict_xgb = xgb.predict(X_test)
print(f"Mean Absolute Error = {str(mean_absolute_error(y_test,predict_xgb))}")
```

```python
import tensorflow as tf
```

```python
ann_model = tf.keras.models.Sequential()
```

```python
ann_model.add(tf.keras.layers.Dense(units = 64,activation="relu"))
ann_model.add(tf.keras.layers.Dropout((0.5)))

ann_model.add(tf.keras.layers.Dense(units = 64,activation="relu"))
ann_model.add(tf.keras.layers.Dropout((0.5)))
              

              
ann_model.add(tf.keras.layers.Dense(units = 1,activation="linear"))

ann_model.compile(optimizer="adam",loss="mean_squared_error",metrics=["mae"])
ann_model.fit(X_train,y_train,epochs=25,batch_size=32,validation_data=(X_test,y_test))
              
```
